#### Basic chunking example

```python
from llama_index.core.node_parser import SentenceWindowNodeParser
from llama_index.core import SimpleDirectoryReader, Document

node_parser = SentenceWindowNodeParser.from_defaults(
    window_size=2,
    window_metadata_key="window",
    original_text_metadata_key="original_text",
)

text = "I love programming. Python is my most favorite language. I love LLMs. I love LlamaIndex."

nodes = node_parser.get_nodes_from_documents([Document(text=text)])

print([x.text for x in nodes])

print(nodes[1].metadata["window"])
```

```python
documents = SimpleDirectoryReader("./my_data/").load_data(show_progress=True)

document = Document(text="\n\n".join([doc.text for doc in documents]))

print(document.text[:1000])
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
def split_into_windows(text, window_size=5, overlap=2):
    sentences = text.split('. ')
    windows = []
    for i in range(0, len(sentences), window_size - overlap):
        window = '. '.join(sentences[i:i+window_size])
        if window:
            windows.append(window)
    return windows
```

```python
sentence_windows = split_into_windows(document.text)
```

```python
embeddings = model.encode(sentence_windows)
```

```python
from pinecone import Pinecone, ServerlessSpec
from google.colab import userdata

# initialize connection to pinecone
pinecone = Pinecone(
        api_key=userdata.get("PINECONE_API_KEY")
)

# create the index if it does not exist already
index_name = 'sentence-window-index'

# connect to the index
index = pinecone.Index(index_name)
```

```python
batch_size = 100  # Pinecone recommends batching upserts
for i in range(0, len(embeddings), batch_size):
    batch_embeddings = embeddings[i:i+batch_size]
    ids = [str(j) for j in range(i, i + len(batch_embeddings))]  # Create unique IDs for each window
    metadata = [{'text': sentence_windows[j]} for j in range(i, i + len(batch_embeddings))]
    index.upsert(vectors=list(zip(ids, batch_embeddings, metadata)))

print(f"Upserted {len(embeddings)} sentence windows into Pinecone.")
```

```python
# Function to retrieve relevant windows based on a query
def retrieve_windows(query, top_k=3):
    # Encode the query into an embedding
    query_embedding = model.encode([query])

    # Perform the similarity search in Pinecone
    result = index.query(vector=query_embedding.tolist(), top_k=top_k, include_metadata=True)

    # Return the retrieved sentence windows
    return [match['metadata']['text'] for match in result['matches']]
```

```python
query = "History of tea in China"
results = retrieve_windows(query, top_k=3)

# Print the retrieved windows
for i, result in enumerate(results):
    print(f"Result {i+1}: {result}\n")
```

```python
B_INST, E_INST = "[INST]", "[/INST]"
B_SYS, E_SYS = "<<SYS>>\n", "\n<</SYS>>\n\n"
DEFAULT_SYSTEM_PROMPT = """\
  You will be given texts related to a certain topic. Write a summary response that answers the question based on what is discussed in the texts.
  Do not mention anything outside of what is provided. Don't answer anything outside the context you are provided.
  If there isn't enough context, simply reply "This topic was not discussed previously"
  """

SYSTEM_PROMPT = B_SYS + DEFAULT_SYSTEM_PROMPT + E_SYS

def get_prompt(instruction):
    prompt_template =  B_INST + SYSTEM_PROMPT + instruction + E_INST
    return prompt_template
```

```python
def format_prompt(query, context):
    return '''
    ### Texts:
    {context}

    ### Question:
    {query}
    '''.format(context=context, query=query)
```

```python
from together import Together

client = Together(api_key="c7021769ffb6f26d17ddf6a3abbc56155f6f1165eb039fd0753af0196cdd705c")

# Define a function that runs the model
def answer(query):
    context = retrieve_windows(query, top_k=3)

    prompt = format_prompt(query, context)
    prompt_template = get_prompt(prompt)

    stream = client.chat.completions.create(
    model="mistralai/Mixtral-8x7B-Instruct-v0.1",
    messages=[{"role": "user", "content": prompt_template}],
    stream=True,
    )

    for chunk in stream:
        print(chunk.choices[0].delta.content or "", end="", flush=True)
```
