```python
import os
from llama_index.llms.gemini import Gemini
from llama_index.embeddings.gemini import GeminiEmbedding
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core.node_parser import SentenceWindowNodeParser
from llama_index.core.node_parser import SentenceSplitter
```

```python
from llama_index.core import Settings
```

### Create sentence window node parser with default settings

```python
node_parser = SentenceWindowNodeParser.from_defaults(
    window_size=3,
    window_metadata_key="window",
    original_text_metadata_key="original_text",
)
```

#### I am using Gemini-1.5-flash-latest model and HuggingFace's all-mpnet-base-v2 model for creating embeddings
I removed my api_key, to test things you can put your api key in the empty field below

```python
text_splitter = SentenceSplitter()

llm = Gemini(model="models/gemini-1.5-flash-latest", api_key="")
embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-mpnet-base-v2", max_length=512
)
```

```python
Settings.llm = llm
Settings.embed_model = embed_model
Settings.text_splitter = text_splitter
```

#### Here I have created a big text telling a story about Atlantis

```python
text = "Atlantis, introduced by Plato in his dialogues Timaeus and Critias, is a mythical island that symbolizes the dangers of hubris and imperial overreach. Described as a powerful naval empire, Atlantis ruled much of the known Western world, with vast wealth, technological sophistication, and a military prowess unmatched by any other civilization. However, Plato uses it as a moral foil to 'Ancient Athens,' which he portrays as embodying the philosophical virtues of his ideal state outlined in The Republic. While Atlantis represents material success and power, Athens is the symbol of wisdom, moderation, and civic virtue. \n The myth is a cautionary tale about the dangers of arrogance and unchecked ambition. After attempting to conquer Athens, Atlantis is punished by the gods, leading to its dramatic submersion into the Atlantic Ocean. This downfall serves as a divine response to the Atlanteans' overconfidence, moral corruption, and belief in their invincibility. Through this story, Plato critiques imperialism and warns that even the most powerful civilizations, no matter how advanced, are vulnerable to collapse when they lose their ethical foundation. The Atlantis myth reflects not just Plato’s philosophy but also his critique of the political climate of his time, particularly the imperialistic ambitions of Athens during the Peloponnesian War. By contrasting a corrupt and overreaching Atlantis with a virtuous and disciplined Athens, Plato illustrates the fragility of empires built on power rather than virtue, offering a timeless lesson about governance, ambition, and the consequences of moral failure, with a lighthearted reminder to not let pride sink your ship."
```

#### Create a document from the above text

```python
from llama_index.core.node_parser import SentenceWindowNodeParser
from llama_index.core import SimpleDirectoryReader, Document

docs = [Document(text=text)]
```

### Extract Nodes

I'll extract out the set of nodes that will be stored in the VectorIndex. This includes both the nodes with the sentence window parser, as well as the "base" nodes extracted using the standard parser.

```python
nodes = node_parser.get_nodes_from_documents(docs)
```

```python
base_nodes = text_splitter.get_nodes_from_documents(docs)
```

```python
len(nodes)
```

### Build the Indexes

I'll build both the sentence index, as well as the "base" index (with default chunk sizes).

```python
from llama_index.core import VectorStoreIndex
```

```python
sentence_index = VectorStoreIndex(nodes)

base_index = VectorStoreIndex(base_nodes)
```

### Querying

Here, I'll now use the *MetadataReplacementPostProcessor* to replace the sentence in each node with it's surrounding context.

```python
from llama_index.core.postprocessor import MetadataReplacementPostProcessor
```

```python
query_engine = sentence_index.as_query_engine(
    similarity_top_k=2,
    node_postprocessors=[
        MetadataReplacementPostProcessor(target_metadata_key="window")
    ],
)
window_response = query_engine.query(
    "What is the main difference between Atlantis and Ancient Athens in Plato's portrayal?",
)
print(window_response)
```

I can also check the original sentence that was retrieved for each node, as well as the actual window of sentences that was sent to the LLM.

```python
window = window_response.source_nodes[0].node.metadata["window"]
sentence = window_response.source_nodes[0].node.metadata["original_text"]

print(f"Window: {window}")
print("------------------")
print(f"Original Sentence: {sentence}")
```

### Contrast with normal VectorStoreIndex

```python
query_engine = base_index.as_query_engine(similarity_top_k=3)
vector_response = query_engine.query(
    "What is the main difference between Atlantis and Ancient Athens in Plato's portrayal?"
)
print(vector_response)
```

### Evaluation

```python
eval_dataset = QueryResponseDataset.from_json("atlantis_eval_qr_dataset.json")
```

## Compare Results

```python
from llama_index.core.evaluation import (
    CorrectnessEvaluator,
    SemanticSimilarityEvaluator,
    RelevancyEvaluator,
    FaithfulnessEvaluator,
    PairwiseComparisonEvaluator,
)
```

```python
from collections import defaultdict
import pandas as pd
```

```python
evaluator_c = CorrectnessEvaluator(llm=llm)
evaluator_s = SemanticSimilarityEvaluator()
evaluator_r = RelevancyEvaluator(llm=llm)
evaluator_f = FaithfulnessEvaluator(llm=llm)
```

```python
from llama_index.core.evaluation.eval_utils import (
    get_responses,
    get_results_df,
)
from llama_index.core.evaluation import BatchEvalRunner
```

```python
max_samples = 1

eval_qs = eval_dataset.questions
ref_response_strs = [r for (_, r) in eval_dataset.qr_pairs]

base_query_engine = base_index.as_query_engine(similarity_top_k=2)

query_engine = sentence_index.as_query_engine(
    similarity_top_k=2,
    node_postprocessors=[
        MetadataReplacementPostProcessor(target_metadata_key="window")
    ],
)
```

```python
import numpy as np

base_pred_responses = get_responses(
    eval_qs[:max_samples], base_query_engine, show_progress=True
)
pred_responses = get_responses(
    eval_qs[:max_samples], query_engine, show_progress=True
)

pred_response_strs = [str(p) for p in pred_responses]
base_pred_response_strs = [str(p) for p in base_pred_responses]
```

```python
evaluator_dict = {
    "correctness": evaluator_c,
    "faithfulness": evaluator_f,
    "relevancy": evaluator_r,
    "semantic_similarity": evaluator_s,
}
batch_runner = BatchEvalRunner(evaluator_dict, workers=2, show_progress=True)
```

```python
eval_results = await batch_runner.aevaluate_responses(
    queries=eval_qs[:max_samples],
    responses=pred_responses[:max_samples],
    reference=ref_response_strs[:max_samples],
)
```

```python
base_eval_results = await batch_runner.aevaluate_responses(
    queries=eval_qs[:max_samples],
    responses=base_pred_responses[:max_samples],
    reference=ref_response_strs[:max_samples],
)
```

```python
results_df = get_results_df(
    [eval_results, base_eval_results],
    ["Sentence Window Retriever", "Base Retriever"],
    ["correctness", "relevancy", "faithfulness", "semantic_similarity"],
)
display(results_df)
```
