This function will run the query and will return the data which is returned from neo4j database. Since the data returned is in natural language fromat and returned as a sentence, I aliased it with 'sentence' and used to get the sentence out of the returned object

```python
def run_query(query, parameters=None):
  with driver.session() as session:
    results = session.run(query, parameters)
    return [str(result['sentence']) for result in results]
```

The following function will return what exactly returned from the neo4j database when given a query

```python
def return_from_db(query, parameters=None):
  with driver.session() as session:
    results = session.run(query, parameters)
    return [result for result in results]
```

The following function will match a user query with a pinecone embeddings and list get the one with highest score. I am using an score of 0.5 only because since I haven't joined all the texts and then splitted again with the same size and overlapping factor, I wasn't able to overlap data which will be good to increase scores of matches. But overlapping isn't needed since I didn't split in unwanted position the score it a little bit low evenif the returned value was correct.

```python
def get_data_from_db(query):
    xq = embed_model.get_text_embedding(query)
    result = pinecone_index.query(vector=xq, top_k=20, includeMetadata=True)
    matches = []
    for i in result['matches']:
      if (i['score'] > 0.5):
        # print(i.metadata['text'], " -> score = ", i['score'])
        # append best results
        matches.append(i.metadata['text'])
    return str(matches)
```

```python
B_INST, E_INST = "[INST]", "[/INST]"
B_SYS, E_SYS = "<<SYS>>\n", "\n<</SYS>>\n\n"
DEFAULT_SYSTEM_PROMPT = """\
  You will be given texts related to a certain topic. Write a summary response that answers the question based on what is discussed in the texts.
  Do not mention anything outside of what is provided. Don't answer anything outside the context you are provided.
  If there isn't enough context, simply reply "This topic was not discussed previously and do not say in the text provided or do not mention there is a text provided"
  """

SYSTEM_PROMPT = B_SYS + DEFAULT_SYSTEM_PROMPT + E_SYS

def get_prompt(instruction):
    prompt_template =  B_INST + SYSTEM_PROMPT + instruction + E_INST
    return prompt_template
```

```python
def format_prompt(query, context):
    return '''
    ### Texts:
    {context}

    ### Question:
    {query}
    '''.format(context=context, query=query)
```

This function will convert the user prompt to a neo4j query which will then be used to fetch from the database and when the query returns an error it will simply display all the values present inside the database later.

```python
def nl_to_neoquery(userPrompt):
  import re
  prompt = "give me a cypher query which will eb put directly into a neo4j database for the following natural language query, use PLAYER, COACH, and TEAM in capital letter FOR NODES, return only the query in square bracket([ and ]), and do not comment on anything else and do not mention anything else just return the text, not code: "
  stream = client.chat.completions.create(
    model="mistralai/Mixtral-8x7B-Instruct-v0.1",
    messages=[{"role": "user", "content": prompt + userPrompt}],
    stream=True,
  )
  query = ""
  for chunk in stream:
    query += chunk.choices[0].delta.content or ""

  matches = re.findall(r'\[.*?\]', query, re.DOTALL)
  clean_text = matches[0][1:-1]

  return clean_text
```

```python
def answer(query):
    context = get_data_from_db(query)

    prompt = format_prompt(query, context)
    prompt_template = get_prompt(prompt)
    neoquery = nl_to_neoquery(query)

    stream = client.chat.completions.create(
        model="mistralai/Mixtral-8x7B-Instruct-v0.1",
        messages=[{"role": "user", "content": prompt_template}],
        stream=True,
    )

    for chunk in stream:
        print(chunk.choices[0].delta.content or "", end="", flush=True)

    print("\n\nFrom Neo4j database")
    try:
      result = return_from_db(neoquery)
    except Exception as e:
      print(f"An error occurred: {e}")
      result = return_from_db("MATCH (n) RETURN n")

    print(result)
```
