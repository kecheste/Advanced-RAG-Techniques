#### Here I have created a big text telling a story about Atlantis

```python
doc_text = "Atlantis, introduced by Plato in his dialogues Timaeus and Critias, is a mythical island that symbolizes the dangers of hubris and imperial overreach. Described as a powerful naval empire, Atlantis ruled much of the known Western world, with vast wealth, technological sophistication, and a military prowess unmatched by any other civilization. However, Plato uses it as a moral foil to 'Ancient Athens,' which he portrays as embodying the philosophical virtues of his ideal state outlined in The Republic. While Atlantis represents material success and power, Athens is the symbol of wisdom, moderation, and civic virtue. \n The myth is a cautionary tale about the dangers of arrogance and unchecked ambition. After attempting to conquer Athens, Atlantis is punished by the gods, leading to its dramatic submersion into the Atlantic Ocean. This downfall serves as a divine response to the Atlanteans' overconfidence, moral corruption, and belief in their invincibility. Through this story, Plato critiques imperialism and warns that even the most powerful civilizations, no matter how advanced, are vulnerable to collapse when they lose their ethical foundation. The Atlantis myth reflects not just Plato’s philosophy but also his critique of the political climate of his time, particularly the imperialistic ambitions of Athens during the Peloponnesian War. By contrasting a corrupt and overreaching Atlantis with a virtuous and disciplined Athens, Plato illustrates the fragility of empires built on power rather than virtue, offering a timeless lesson about governance, ambition, and the consequences of moral failure, with a lighthearted reminder to not let pride sink your ship."
```

### Create a document from the above text

```python
from llama_index.core import Document

docs = [Document(text=doc_text)]
```

```python
from llama_index.core.node_parser import (
    HierarchicalNodeParser,
    SentenceSplitter,
)
```

### Extracting nodes from the above document

```python
node_parser = HierarchicalNodeParser.from_defaults()

nodes = node_parser.get_nodes_from_documents(docs)

len(nodes)
```

```python
from llama_index.core.node_parser import get_leaf_nodes, get_root_nodes
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
```

#### *Leaf node* is a node with no child / the high level node

```python
leaf_nodes = get_leaf_nodes(nodes)

len(leaf_nodes)

root_nodes = get_root_nodes(nodes)
```

## Load into Storage

We define a docstore, which we load all nodes into.
We then define a VectorStoreIndex containing just the leaf-level nodes.

Put your gemini api_key in the empty field

```python
from llama_index.core.storage.docstore import SimpleDocumentStore
from llama_index.core import StorageContext
from llama_index.llms.gemini import Gemini
 
docstore = SimpleDocumentStore()

docstore.add_documents(nodes)

storage_context = StorageContext.from_defaults(docstore=docstore)

llm = Gemini(model="models/gemini-1.5-flash-latest", api_key="")
```

```python
from llama_index.core import Settings

embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-mpnet-base-v2", max_length=512
)

Settings.llm = llm
Settings.embed_model = embed_model
```

```python
from llama_index.core import VectorStoreIndex

base_index = VectorStoreIndex(
    leaf_nodes,
    storage_context=storage_context,
)
```

## Define Retriever

```python
from llama_index.core.retrievers import AutoMergingRetriever
```

```python
base_retriever = base_index.as_retriever(similarity_top_k=6)
retriever = AutoMergingRetriever(base_retriever, storage_context, verbose=True)
```

```python
query_str = (
    "What is the main difference between Atlantis and Ancient Athens in Plato's portrayal?"
)

nodes = retriever.retrieve(query_str)
base_nodes = base_retriever.retrieve(query_str)
```

```python
len(nodes)
```

```python
len(base_nodes)
```

```python
from llama_index.core.response.notebook_utils import display_source_node

for node in nodes:
    display_source_node(node, source_length=1000)
```

## Plug it into Query Engine

```python
from llama_index.core.query_engine import RetrieverQueryEngine
```

```python
query_engine = RetrieverQueryEngine.from_args(retriever)
base_query_engine = RetrieverQueryEngine.from_args(base_retriever)
```

```python
response = query_engine.query(query_str)
```

```python
print(str(response))
```

```python
base_response = base_query_engine.query(query_str)
```

```python
from llama_index.core.evaluation import DatasetGenerator, QueryResponseDataset
import nest_asyncio

nest_asyncio.apply()
```

```python
eval_llm = Gemini(model="models/gemini-1.5-flash-latest", api_key="AIzaSyAp9zJhry-MaIJG7u1fkvBQ37cOWMUf5_g")
dataset_generator = DatasetGenerator(
    root_nodes[:4],
    llm=eval_llm,
    show_progress=True,
    num_questions_per_chunk=2,
)
```

```python
eval_dataset.save_json("atlantis_eval_qr_dataset_AM.json")
```

```python
eval_dataset = QueryResponseDataset.from_json(
    "atlantis_eval_qr_dataset_AM.json"
)
```

```python
from llama_index.core.evaluation import (
    CorrectnessEvaluator,
    SemanticSimilarityEvaluator,
    RelevancyEvaluator,
    FaithfulnessEvaluator,
    PairwiseComparisonEvaluator,
)

from collections import defaultdict
import pandas as pd

evaluator_c = CorrectnessEvaluator(llm=eval_llm)
evaluator_s = SemanticSimilarityEvaluator()
evaluator_r = RelevancyEvaluator(llm=eval_llm)
evaluator_f = FaithfulnessEvaluator(llm=eval_llm)
```

```python
from llama_index.core.evaluation.eval_utils import (
    get_responses,
    get_results_df,
)
from llama_index.core.evaluation import BatchEvalRunner
```

```python
eval_qs = eval_dataset.questions
qr_pairs = eval_dataset.qr_pairs
ref_response_strs = [r for (_, r) in qr_pairs]
```

```python
pred_responses = get_responses(eval_qs, query_engine, show_progress=True)
```

```python
base_pred_responses = get_responses(
    eval_qs, base_query_engine, show_progress=True
)
```

```python
import numpy as np

pred_response_strs = [str(p) for p in pred_responses]
base_pred_response_strs = [str(p) for p in base_pred_responses]
```

```python
evaluator_dict = {
    "correctness": evaluator_c,
    "faithfulness": evaluator_f,
    "relevancy": evaluator_r,
    "semantic_similarity": evaluator_s,
}
batch_runner = BatchEvalRunner(evaluator_dict, workers=2, show_progress=True)
```

```python
eval_results = await batch_runner.aevaluate_responses(
    eval_qs, responses=pred_responses, reference=ref_response_strs
)
```

```python
base_eval_results = await batch_runner.aevaluate_responses(
    eval_qs, responses=base_pred_responses, reference=ref_response_strs
)
```

```python
results_df = get_results_df(
    [eval_results, base_eval_results],
    ["Auto Merging Retriever", "Base Retriever"],
    ["correctness", "relevancy", "faithfulness", "semantic_similarity"],
)
display(results_df)
```
